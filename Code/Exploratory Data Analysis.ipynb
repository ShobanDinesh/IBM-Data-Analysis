```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

It is important to understand what type of variable we are dealing, before visualize individual variables.

```python
df.dtypes
```

```python
df.corr()
```

```python
df[['bore','stroke','compression-ratio','horsepower']].corr()
```

#### Continous numerical varaibles 

```python
sns.regplot(x='engine-size',y='price',data=df)
plt.ylim(0,)
```

```python
df[['engine-size','price']].corr()
```

It can be noted that engine size and price have a positive linear relationship, as the engine size increase, so does the price. The correlation between the variables are 0.87. Engine size can be a great predictor of price 

```python
sns.regplot(x='highway-mpg',y='price',data=df)
```

```python
df[['highway-mpg', 'price']].corr()
```

Highway mpg and price have an inverse/negative relatioship, the correlation coefficient is -0.7. Highway mpg could be a potential predictor of price 

```python
sns.regplot(x='peak-rpm',y='price',data=df)
```

```python
df[['peak-rpm','price']].corr()
```

It is evident that the peak rpm and price have weak linear relationship with correlation coefficient of -0.1, it is not a great predictor of price

```python
sns.regplot(x='stroke',y='price',data=df)
```

```python
df[['stroke','price']].corr()
```
