#### Correlation and Causation 
Correlation is a measure of extent of interdependence of variables

Causation is the relationship of cause and effect between two variables

"Correlation does not imply Causation"

Determining Correlation is much easier then determining causation, as causation might require independent experiment

Pearson's Correlation measures the linear dependence between two variables. It varies from -1 to 1, where -1 represents total negative correlation, 0 represents no linear correlation, 1 represents total positive correlation 

```python
df.corr()
```

P-value: P-value is the probability value that the correlation between 2 variables is statistically significant. Normally, we choose the significance value of 0.05, which means we are 95% confident that the correlation is significant.
By convention, when the 
<ul>
    <li>p-value < 0.001, we can say that the strong evidence that the correlation is significant </li>
    <li>p-value < 0.05, we can say that the moderate evidence that the correlation is significant </li>
    <li>p-value < 0.1, we can say that the weak evidence that the correlation is significant </li>
    <li>p-value > 0.1, we can say that the no evidence that the correlation is significant </li>
</ul>

```python
def corr_p(col_1,col_2='price'):
    p_cor, p_value = stats.pearsonr(df[col_1],df[col_2])
    evidence = ''
    correlation = ''
    if p_value<0.001:
        evidence = 'strong'
    elif p_value<0.05:
        evidence = 'moderate'
    elif p_value<0.1:
        evidence = 'weak'
    else:
        evidence = 'no'
    print(p_cor,'correlation with',evidence,'evidence')
```

#### ANOVA
The Analysis of Variance is a statistical method used to test if there are significant difference between the mean of 2 or more groups. ANOVA returns 2 parameters:
<ul>
    <li>F test score: ANOVA assumes that the means of all groups are the same, calculated how the actual mean deviates from the assumption and reports it as F test score. A larger score means there is larger difference between them.</li>
    <li>P-value tells how statistically significant the calculated value is.</li>
</ul>
If price is strongly correlated with a variable, we can expect ANOVA to return sizeable F test score and small p value 

```python
grouped_df = df[['drive-wheels','price']].groupby(['drive-wheels'])
```

```python
f_val, p_val = stats.f_oneway(grouped_df.get_group('fwd')['price'],grouped_df.get_group('rwd')['price'],grouped_df.get_group('4wd')['price'])
```

```python
f_val,p_val
```

This is a great result, with a large F test score showing a strong correlation and a P value of almost 0 implying almost certain statistical significance. But does this mean all three tested groups are all this highly correlated? 

```python
f_val, p_val = stats.f_oneway(grouped_df.get_group('fwd')['price'], grouped_df.get_group('rwd')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

```python
f_val, p_val = stats.f_oneway(grouped_df.get_group('4wd')['price'], grouped_df.get_group('rwd')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(grouped_df.get_group('4wd')['price'], grouped_df.get_group('fwd')['price'])  
print("ANOVA results: F=", f_val, ", P =", p_val)   
```
